```python
labels=["558","559","561","565","567","586","587","591","592","594","595","597","598","602","604","605","606",
        "607","610","613","614","616","618","622","624","625","626","627","628"]
```

```python
def generate_images(file_names,labels,band):
    k=0
    for i,file in enumerate(file_names):
        mat = open(r'C:/Users/ayush/OneDrive/Desktop/Coherence/Coherence matrices/' + band + '/' + file)
        fcm = np.loadtxt(mat, delimiter=",")
        UT=np.triu([fcm])[0]             ##  Extracting the upper triangle from a functional connectivity matrix 
        
        l=[]                                                       
        for i in range(60):
            l.append(UT[i][i+1:])

        l1=[]
        for i in range(60):                                         ##  Saving the datapoints in the form of a linear array
            for j in range(len(l[i])):
                l1.append(l[i][j])
                
        l2=[]
        l2=l1[0:1764]                                               ## dropping the last six datapoints and converting the linear array into a 42*42 square matrix
        x1 = np.reshape(l2, (42, 42))  
        s=str(labels[k])+"NEW_coherence_"+band+".jpeg"
        plt.imshow(x1)
        fig1 = plt.gcf()
        fig1.savefig(s, dpi=50)

        k=k+1
        
    return
```

```python
s1="ALPHA"
generate_images(csv_files1,labels,s1)
```

```python
s2="BETA"
generate_images(csv_files2,labels,s2)
```

```python
s3="DELTA"
generate_images(csv_files3,labels,s3)
```
